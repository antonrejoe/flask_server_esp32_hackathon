```python
import numpy as np

def gs_orth(sig):
    # Number of columns in the matrix
    n = sig.shape[1]

    for j in range(n):
        # Orthogonalize the vector in column j with respect to the previous vectors
        for k in range(j):
            proj = np.dot(sig[:, k], sig[:, j]) / np.dot(sig[:, k], sig[:, k])
            sig[:, j] = sig[:, j] - proj * sig[:, k]
        norm = np.linalg.norm(sig[:, j])
        if norm != 0:
            sig[:, j] = sig[:, j] / norm
    
    return sig

# Example usage:
A = np.array([[1,1,1,1,1,1,-1,-1]])
orthogonalized_A = gs_orth(A.copy())
print("Orthogonalized matrix:\n", orthogonalized_A)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
T1 = 8  # Total time duration
T2 = 4
N = 1000  # Number of samples
t1 = np.linspace(0, T1, N)
t2 = np.linspace(0, T2, N)

# Define the signals based on the diagram
s1 = np.piecewise(t1, [t1 < 6*T1/8, t1 >= 6*T1/8 ], [5, -5])
s2 = np.piecewise(t2, [t2 < T2/4, (t2 >= T2/4) & (t2 < 3*T2/4), (t2 >= 3*T2/4) & (t2 < 4*T2/4)], [-5, 5, -5])

# Gram-Schmidt orthogonalization
def gram_schmidt(vectors):
    orthonormal_basis = []
    for v in vectors:
        w = v - sum(np.dot(v, b) * b for b in orthonormal_basis)
        if np.linalg.norm(w) > 1e-10:  # Avoid division by zero
            orthonormal_basis.append(w / np.linalg.norm(w))
    return np.array(orthonormal_basis)

# Stack the signals
signals = np.vstack([s1, s2])

# Perform Gram-Schmidt orthogonalization
orthonormal_basis = gram_schmidt(signals)

# Plot the signals and the orthonormal basis functions in the same figure
plt.figure(figsize=(10,6))
plt.subplot(2, 2, 1)
plt.plot(t1, s1, label=r'$s_1(t)$')
plt.grid()
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(t2, s2, label=r'$s_2(t)$')
plt.legend()
plt.grid()

for i, basis in enumerate(orthonormal_basis, start=1):
    plt.subplot(2, 2, i+2)
    plt.plot(t1, basis, label=r'$\phi_{}(t)$'.format(i))
    plt.grid()
    plt.legend()

plt.tight_layout()
plt.show()

# Print the orthonormal basis
print("Orthonormal basis functions:")
for i, basis in enumerate(orthonormal_basis, start=1):
    print(f"\nphi_{i}(t):")
    print(basis)
```
